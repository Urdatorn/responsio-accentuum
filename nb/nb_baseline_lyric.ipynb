To avoid having to recompute the stats for all 337,200 baseline lines if the notebook is reset or we want to calculate the baseline for a specific ode, or set of odes or lines, we save the results as a pickle as we calculate it.

```python
import pickle
import os
from src.stats_comp import compatibility_corpus, compatibility_ratios_to_stats

# Define the pickle file path
pickle_file = 'data/all_comp_ratios_bl_lyric.pkl'

# Check if pickle file exists, if so load it, otherwise compute and save
if os.path.exists(pickle_file):
    print("Loading from pickle file...")
    with open(pickle_file, 'rb') as f:
        all_comp_ratios_bl_lyric = pickle.load(f)
    print("Loaded from pickle file.")
else:
    print("Computing compatibility corpus...")
    all_comp_ratios_bl_lyric = compatibility_corpus('data/compiled/baselines/triads/lyric')
    print("Ratio done.")
    
    # Save to pickle file
    with open(pickle_file, 'wb') as f:
        pickle.dump(all_comp_ratios_bl_lyric, f)
    print(f"Saved to {pickle_file}")

corpus_comp_stat_bl = compatibility_ratios_to_stats(all_comp_ratios_bl_lyric)
print(corpus_comp_stat_bl)
```
